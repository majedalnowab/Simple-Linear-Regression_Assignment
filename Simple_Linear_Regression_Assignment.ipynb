## ***Simple Linear Regression Model:***

```python
yd = df[['year']]
id = df[['per capita income (US$)']]
```

```python
std = sc()
years = std.fit_transform(yd)
income = std.fit_transform(id)
```

```python
years = years.reshape(-1, 1)
model = LinearRegression()
model.fit(years, income)
predicted_income = model.predict(years)

plt.scatter(years, income, label='Actual Data')
plt.plot(years, predicted_income, color='red', label='Linear Regression')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Simple Linear Regression Model')
plt.legend()
plt.show()

print('Intercept:', model.intercept_)
print('Coefficient:', model.coef_[0])
```
